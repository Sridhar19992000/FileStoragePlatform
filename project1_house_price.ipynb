```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1) Fake data: size (sqft) -> price (in thousands)
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # feature
y = np.array([100,  150,  200,  250,  300])                  # label (k$)

# 2) Train the model (computer finds m & b)
model = LinearRegression()
model.fit(X, y)

# 3) Predict one value
pred_for_2200 = model.predict([[2200]])[0]
print(f"Predicted price for 2200 sqft: {pred_for_2200:.1f}k USD")

# 4) Draw points (red) and best-fit line (blue)
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("House size (sqft)")
plt.ylabel("Price (thousands USD)")
plt.title("Project 1: Linear Regression (House Price)")
plt.show()
```
